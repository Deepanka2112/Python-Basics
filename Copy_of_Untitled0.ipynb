2. What is an interpreter in Python?

answer - In Python, an interpreter is a program that reads Python code and executes it line by line, translating it into machine-readable bytecode before execution. Unlike compiled languages, Python is interpreted, meaning the code is not entirely converted to machine code beforehand.

3.  What are pre-defined keywords in Python?

answer: In Python, pre-defined keywords are reserved words with specific meanings used in the language's syntax. They cannot be used as variable names, function names, or class names, and they have special roles in how the Python interpreter executes code.

4.  Can keywords be used as variable names?

Answer: No, keywords cannot be used as variable names in most programming languages. Keywords are reserved words that have specific meanings within the language and are used to define the structure and syntax of the code. Using a keyword as a variable name would lead to a conflict because the interpreter would not know whether the word refers to the keyword's intended meaning or the variable.

5.  What is mutability in Python?

Answer: In Python, mutability refers to an object's ability to be changed after it's been created. Mutable objects, like lists and dictionaries, can have their contents altered without creating a new object, while immutable objects, like strings and tuples, cannot be modified once they're created.

6. 8 Why are lists mutable, but tuples are immutable?

Answer - In Python, lists and tuples both store collections of data, but differ in mutability, performance and memory usage. Lists are mutable, allowing modifications, while tuples are immutable. Choosing between them depends on whether you need to modify the data or prioritize performance and memory efficiency

7.  What is the difference between “==” and “is” operators in Python?

Answer: In Python, == and is are both comparison operators, but they compare different aspects of objects. == checks for equality of values, meaning it returns True if the two objects have the same content. is checks for identity, meaning it returns True only if the two variables refer to the same object in memory.

8.  What are logical operators in Python?
Answer: In Python, logical operators are used to combine conditional statements and expressions, returning a Boolean result (True or False). They are used for making decisions and controlling the flow of your program based on the truth value of expressions.

9. What is type casting in Python?

Answer: Type casting, also known as type conversion, is the process of changing the data type of a variable in Python. It's useful for performing operations on different types of data and ensuring compatibility between them.

10. 8 What is the difference between implicit and explicit type casting?

Answer: Implicit type casting, also known as widening casting, automatically converts a smaller data type to a larger one, without the need for explicit instruction from the programmer. Explicit type casting, or narrowing casting, requires the programmer to manually specify the conversion of a larger data type to a smaller one.

11.  What is the purpose of conditional statements in Python?

Answer: Conditional statements in Python (like if, elif, and else) allow a program to make decisions and execute different code blocks based on whether a condition is true or false. They control the flow of execution, enabling dynamic and flexible program behavior.

12.  How does the elif statement work?

Answer: The “elif” keyword in Python, stands for “else if”. It can be used in conditional statements to check for multiple conditions. For example, if the first condition is false, it moves on to the next “elif” statement to check if that condition is true.
